```python
import pandas as pd
data = pd.read_csv("C://Users//midla//Downloads//student_performance.csv")
data.head()
```

```python
import matplotlib.pyplot as plt
x=data['Name']
y=data['Math_Score']
plt.bar(x,y,color="green")
plt.xlabel("STUDENT NAME")
plt.ylabel("MATHS SCORE")

plt.show()
```

```python
x=data['Name']
y1=data['Writing_Score']
y2=data['Reading_Score']
plt.plot(x,y1,color="red",label="Writing_Score")
plt.plot(x,y2,color="green",label="Reading_Score")
plt.xlabel("NAME")
plt.ylabel("SCORE")

plt.title("READING VS WRITING")
plt.legend()
plt.show()
```

```python
import plotly.express as px
```

```python
fig=px.bar(data,x='Name',y='Writing_Score')

fig.show()
```

```python
gender_counts = data['Gender'].value_counts().reset_index()
gender_counts.columns = ['Gender', 'Count']
fig=px.pie(data,values='Count',names='Gender')
fig.show()
```
